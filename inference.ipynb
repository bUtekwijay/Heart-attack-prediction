```python
with open('final_best_model.pkl', 'rb') as file:
  model = pickle.load(file)
```

```python
data_inf = pd.DataFrame([
{
    'age': 45,               # nilai baru untuk usia
    'sex': 'F',                  # nilai baru untuk jenis kelamin
    'chestpaintype': 'NAP',        # nilai baru untuk jenis nyeri dada
    'restingbp': 170,        # nilai baru untuk tekanan darah istirahat
    'cholesterol': 250,      # nilai baru untuk kolesterol
    'fastingbs': 1,            # nilai baru untuk gula darah puasa
    'restingecg': 'LVH',           # nilai baru untuk hasil EKG istirahat
    'maxhr': 160,            # nilai baru untuk detak jantung maksimum
    'exerciseangina': 'Y',       # nilai baru untuk angina akibat olahraga
    'oldpeak': 1.5,            # nilai baru untuk depresi ST
    'st_slope': 'Down'  
},
{
    'age': 60,
    'sex': 'M',
    'chestpaintype': 'ASY',
    'restingbp': 140,
    'cholesterol': 300,
    'fastingbs': 1,
    'restingecg': 'ST',
    'maxhr': 130,
    'exerciseangina': 'Y',
    'oldpeak': 2.3,
    'st_slope': 'Flat'
}])
data_inf
```

```python
# Prediksi batch
predictions = model.predict(data_inf)

# Tampilkan hasil
results = pd.DataFrame({
    'Prediction': ['Heart Disease' if p == 1 else 'No Heart Disease' for p in predictions],
})
print(results)
```
